```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("/content/DigitalAd_dataset.csv")
```

```python
data.shape
```

```python
data.head(5)
```

Segregating Datset into X=input/independent variable ,Y = output/dependent variable



```python
X=data.iloc[:,:-1].values # : represent the all rows ,:-1 represnts the all columns except last colummn
X
```

```python
Y=data.iloc[:,-1].values #-1 represents the last coln
Y
```

spiliiting datset

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
```

Feature Scaling

Fit_Transform - fit method is caluculate the mean and varaince of each of the individual features present in the data


Transform - transform method  features using resepctive mean and variance


```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train =sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

Training

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=0)
model.fit(X_train,Y_train)
```

PREDCTING whether new customer with Age and Salary will Buy or Not

```python
age = int(input("enter the new customer's Age: "))
sal = int(input("enter the new customer's Sal: "))
newcust =[[age,sal]]
result =model.predict(sc.transform(newcust))
print(result)
if result == 1:
  print("customer will buy")
else:
  print("customer wont buy")
```

```python
y_pred = model.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),Y_test.reshape(len(Y_test),1)),1))
```

accuracy : confusion matrix

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(Y_test,y_pred)
print("confusion Matrix: ")
print(cm)
print("accuracy of the model:{0}%".format(accuracy_score(Y_test,y_pred)*100))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
